```python
import pandas as pd
import os
import json
import glob

# Paths
input_csv_path = "/Users/shreya/Documents/UCDavis/Practicum/Spring_Quarter_Practicum/Web_Scraping_MyCodes/shreyaV2/Agency_Names_Maps.csv"
base_json_dir = "/Users/shreya/Documents/UCDavis/Practicum/Spring_Quarter_Practicum/Web_Scraping_MyCodes/shreyaV2/outputs"

# Read the CSV
df = pd.read_csv(input_csv_path)

# Store placeId columns per row as lists
place_id_columns = []

# Track max number of placeIds across all agencies to determine column expansion
max_place_ids = 0

for agency_name in df['agency_name_web_scraped_list']:
    folder_name = agency_name.replace(" ", "_")
    json_files = glob.glob(os.path.join(base_json_dir, folder_name, "maps_*.json"))

    place_ids = []

    if json_files:
        json_path = json_files[0]  # Pick the first matching file
        try:
            with open(json_path, 'r') as f:
                data = json.load(f)
                for place in data.get('places', []):
                    place_ids.append(place.get('placeId'))
        except Exception as e:
            print(f"Error reading {json_path} for {agency_name}: {e}")

    else:
        print(f"No JSON file found for {agency_name}")

    # Track the longest list of placeIds seen so far
    max_place_ids = max(max_place_ids, len(place_ids))

    # Add list of placeIds for this row
    place_id_columns.append(place_ids)

# Add new columns for placeId_1, placeId_2, ..., placeId_n
for i in range(max_place_ids):
    df[f'placeId_{i+1}'] = [place_ids[i] if i < len(place_ids) else None for place_ids in place_id_columns]

# Save updated DataFrame back to CSV
output_csv_path = input_csv_path.replace(".csv", "_with_placeIds.csv")
df.to_csv(output_csv_path, index=False)

print(f" Extracted up to {max_place_ids} placeId(s) per agency and saved to:")
```
